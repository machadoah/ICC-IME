# 1. Repetição com While

> definição do loop ou "laço de repetição"


    while condição:

        comado (que só será executado caso a condição acima for verdadeira)

    comando (que vai ser executado independentemente da condição) 

##### Exemplos:

```python
# (1) Realizar a potencia de 2 de 0 até 10:

print(2 ** 0)
print(2 ** 1)
print(2 ** 2)
print(2 ** 3)
print(2 ** 4)
print(2 ** 5)
print(2 ** 6)
print(2 ** 7)
print(2 ** 8)
print(2 ** 9)
print(2 ** 10)
```

```python
# (2) Realizar a potencia de 2 de 0 até 10:

i = 0

while i <= 10:
    
    print(2 ** i)
    
    i = i + 1
```

```python
# (3) Realize um algoritmo que receba varios numeros e quando se digitar 0, o algoritmo forneça a soma de todos os numeros digitados

print("Digite uma sequencia de valores terminadas em zero")

soma = 0

value = 1

while value != 0:
    
    value = int(input("Insira um numero a ser somado: "))
    
    soma = soma + value
    
print("A soma é: {}".format(soma))
```

```python
# (4) Realize um algoritmo que pergunte ao usuários quantos numeros ele deseja somar, e no final da sequencia digitada forneça o resultado

TotalNumeros = int(input("Forneça a quantidade de numeros que deseja somar: "))

print("Digite uma sequencia de valores terminadas em zero")

contador = 0

soma = 0

value = 0

while TotalNumeros != contador:
    
    value = int(input("Insira um numero a ser somado: "))
    
    soma = soma + value
    
    contador = contador + 1
    
print("A soma é: {}".format(soma))
```

## 1.1. Desafio

> realize um algoritmo que realize a soma dos digitos que compoem um número

```python
cumulativa = 0

restanteNumero = 0

ultimoDigito = 0

numero = int(input("Insira um numero inteiro positivo: "))

while numero != 0:

    ultimoDigito = numero % 10

    restanteNumero = numero // 10
    
    # ----------------------------- 
    
    cumulativa = cumulativa + ultimoDigito

    # -----------------------------

    numero = restanteNumero

print ("a soma dos numeros que compoem o valor inserido por você é: {}".format(cumulativa))
```
