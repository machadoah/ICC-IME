# 2. Execusão Condicional

> if , if ... else, biblioteca math

**IF**

    if condição:

        comado (que só será executado caso a condição acima for verdadeira)

    comando (que vai ser executado independentemente da condição)

```python
temperatura = float(input("Insira uma temperatura em graus celcius(°C): "))

if temperatura >= 100:
    
    print("A água esta em ebulição")
```

**IF ... ELSE**

    if condição:

        comado (que só será executado caso a condição acima for verdadeira)
        
    else:

        comando (que vai ser executado caso a condição seja falsa)

```python
temperatura = float(input("Insira uma temperatura em graus celcius(°C): "))

if temperatura >= 100:
    
    print("A água esta em ebulição")
    
else: 
    
    print("A água ainda não entrou em ebulição")
```

##### Exercicio 1. Realize um algoritmo generico que calcule as raizes de uma equação do 2º grau


```python
import math

a = float(input("insira o a: "))
b = float(input("insira o b: "))
c = float(input("insira o c: "))


delta = b ** 2 - 4 * a * c

x1 = (-b + math.sqrt(delta)) / (2 * a)

x2 = (-b - math.sqrt(delta)) / (2 * a)

if delta == 0:
    
    print (f"a unica raiz é {x1}")
    
else:
    
    if delta < 0:
        
        print("Não existe raiz")
        
    else:
        
        print("a primeira raiz é {}  e a segunda é {}".format(x1,x2))
```
